______________
#### Page 9 - Separating Columns in Python
**Step 1 - option 1: separate the columns**

```python
babies1 = babies['Address'].str.split('/', expand=True)
```

```python
babies1.head()
```

**Step 1 - option 2: rename the columns while splitting**

```python
babies2 = babies['Address'].str.split('/', expand=True).rename(columns = lambda x: 'Address'+str(x+1))
```

```python
babies2.head()
```

**Step 2: add the columns back into dataset**

```python
babies3 = pd.concat([babies, babies2], axis=1)
```

```python
babies3.head()
```

*Note: This dataset is a little ridiculous b/c it includes the originally named separate columns, the merged single columns from those ones, and the re-split version of those with new, less clear names...*
