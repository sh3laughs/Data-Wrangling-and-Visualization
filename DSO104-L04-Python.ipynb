#### Page 11 - Scatter Plots in Python

**Import data**

```python
nutrition = pd.read_csv('/Users/hannah/Library/CloudStorage/GoogleDrive-gracesnouveaux@gmail.com/My Drive/Bethel Tech/Data Science/DSO104 Data Wrangling and Visualization/Lesson 4: Displaying Quantitative Data/USDA_nutrition.csv')
```

```python
nutrition
```

**.plot.scatter() pandas function**

```python
plot = nutrition.plot.scatter(x = 'fats', y = 'sat fats')
plot.set_xlabel('Fats')
plot.set_ylabel('Saturated Fats')
plot.set_title('Fats by Saturated Fats')
```

**Add in third variable with 'c' argument - turns it to greyscale**

```python
plot = nutrition.plot.scatter(x = 'fats', y = 'sat fats', c = 'protein')
plot.set_xlabel('Fats')
plot.set_ylabel('Saturated Fats')
plot.set_title('Fats by Saturated Fats and Protein')
```

**Add 'cmap' argument to make it more colorful**

```python
plot = nutrition.plot.scatter(x = 'fats', y = 'sat fats', c = 'protein',
                             cmap = 'coolwarm')
plot.set_xlabel('Fats')
plot.set_ylabel('Saturated Fats')
plot.set_title('Fats by Saturated Fats and Protein')
```

**Replace 'c' and 'cmap' arguments with an 's' argument for the third variable to be visualized by size instead of color**

```python
plot = nutrition.plot.scatter(x = 'fats', y = 'sat fats', 
                              s = nutrition['protein'])
plot.set_xlabel('Fats')
plot.set_ylabel('Saturated Fats')
plot.set_title('Fats by Saturated Fats and Protein')
```

**Add a multiplier so all dots are easy to see**

```python
plot = nutrition.plot.scatter(x = 'fats', y = 'sat fats', 
                              s = nutrition['protein'] * 2)
plot.set_xlabel('Fats')
plot.set_ylabel('Saturated Fats')
plot.set_title('Fats by Saturated Fats and Protein')
```
